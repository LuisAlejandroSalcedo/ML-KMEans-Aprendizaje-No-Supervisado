# KMeans
KMeans es un algoritmo de aprendizaje no supervisado que agrupo la data segun las caracteristicas que tiene.



```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn import metrics
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
km = KMeans(n_clusters=3, max_iter=3000)
```

```python
km.fit(x)
```

```python
predicciones = km.predict(x)
```

```python
predicciones
```

```python
score = metrics.adjusted_rand_score(y, predicciones)
```
